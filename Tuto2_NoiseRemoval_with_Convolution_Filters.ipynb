# Tuto2 Noise Removal with ImageProcessing Filters

#### 1) Add Gaussian Noises 
#### 2) Mean Filter (Average)
#### 3) Gaussian Blur Filter
#### 4) Median Filter 


```python
##import dependencies
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

img = cv2.imread('eye.jpg')


#OpenCV default format is BGR
#Now We'll convert BGR to RGB
#Remove axis
RGB  = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(RGB)
plt.show
```

```python
#Check shape of input image
RGB = np.array(RGB)
RGB.shape
```

```python
#Convert RGB to Grayscale
gray = cv2.cvtColor(RGB,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

# Add Gaussian Noises

```python
mean = 10
var = 1000
sigma = var**0.5
row,col = 556,1000    #width and height of RGB image =
gauss = np.random.normal(mean,sigma,(row,col))
gauss = gauss.reshape(row,col)
gray_noisy = gray + gauss
gray_noisy = np.array(gray_noisy)
gray_noisy= gray_noisy.astype(np.int16)
plt.imshow(gray_noisy,cmap='gray')
```

# Apply Mean Fitler (Average) on Noisy Image

```python
blur = cv2.blur(gray_noisy,(5,5))

#Plot multiple Images

plt.subplot(221),plt.imshow(gray_noisy,cmap='gray'),plt.title('Noisy_Image')
plt.xticks([]),plt.yticks([])
plt.subplot(222),plt.imshow(blur,cmap='gray'),plt.title('Blurred')
plt.xticks([]),plt.yticks([])
plt.show()
```

# Apply Gaussian Blur Filter on Noisy Image

```python
gblur = cv2.GaussianBlur(gray_noisy,(5,5),0)

#Plot multiple Images

plt.subplot(221),plt.imshow(gray_noisy,cmap='gray'),plt.title('Noisy_Image')
plt.xticks([]),plt.yticks([])
plt.subplot(222),plt.imshow(gblur,cmap='gray'),plt.title('Gaussian_Blurred')
plt.xticks([]),plt.yticks([])
plt.show()
```

# Apply Median Filter on RGB Image

```python
median = cv2.medianBlur(gray_noisy,5)  

#Plot multiple Images

plt.subplot(221),plt.imshow(gray_noisy,cmap='gray'),plt.title('Noisy_Image')
plt.xticks([]),plt.yticks([])
plt.subplot(222),plt.imshow(median,cmap='gray'),plt.title('Median_Filter')
plt.xticks([]),plt.yticks([])
plt.show()

```

```python
#Check Medium Filter on Noisy Image

plt.axis("off")
plt.imshow(median,cmap='gray')
```

```python
#Compare Mean Filter, Gaussian Blurred , Median Filter on Noisy Image

plt.subplot(221),plt.imshow(gray_noisy,cmap='gray'),plt.title('Noisy_Image')
plt.xticks([]),plt.yticks([])
plt.subplot(222),plt.imshow(blur,cmap='gray'),plt.title('Mean Filter')
plt.xticks([]),plt.yticks([])
plt.subplot(223),plt.imshow(gblur,cmap='gray'),plt.title('Gaussian_Blurred')
plt.xticks([]),plt.yticks([])
plt.subplot(224),plt.imshow(median,cmap='gray'),plt.title('Median_Filter')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#Save Image

cv2.imwrite('noise.jpg',gray_noisy)
cv2.imwrite('Mean_Filter.jpg',blur)
cv2.imwrite('Gaussian_Filter.jpg',gblur)
cv2.imwrite('Median_Filter.jpg',median)
```
