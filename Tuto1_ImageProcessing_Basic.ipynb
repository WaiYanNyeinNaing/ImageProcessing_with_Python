# Basic Image Processing Tips

#### 1) Image Read / Display Image
#### 2) BGR to RGB
#### 3) RGB to GrayScale
#### 4) Axis off
#### 5) GrayScale to B&W

```python
#import dependencies
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline


# Image Read
img = cv2.imread('eye.jpg')
plt.imshow(img)
plt.show()
```

```python
#(OpenCV default format is BGR)
# Now We'll convert BGR to RGB
#Remove axis

RGB  = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(RGB)
plt.show
```

```python
#check RGB image shape (Width X Height X number_channels =3)
RGB = np.array(RGB)
print(RGB)
print("Shape of Image is = " ,RGB.shape)
```

```python
#Convert RGB Image to Grayscale

gray = cv2.cvtColor(RGB,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
plt.axis("off")
plt.show()
```

```python
#Check Grayscale shape (Width X Height X number_channels =1)
gray = np.array(gray)
print(gray)
print("Shape of Image is = " ,gray.shape)
```

```python
# Convert Grayscale Image into Black&White (range 0 to 1)

b_w = gray/255
b_w = np.array(b_w)
print(b_w)
print("Shape of Image is = " ,b_w.shape)
```

```python
plt.imshow(b_w,cmap='binary')
plt.axis("off")
plt.show()
```
